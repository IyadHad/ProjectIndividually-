```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('data/titanic_datatset/gender_submission.csv')
df2 = pd.read_csv('data/titanic_datatset/test.csv')
df3 = pd.read_csv('data/titanic_datatset/train.csv')
```

```python
#we delete all the passenger from the dataset when we don't know if they survived or not
let =False
liste = []
for i in df2['PassengerId']:
    for j in df1['PassengerId']:
        if i == j:
            let = True
    if let == False:
        liste.append(i)
    let=False
indexnames = []
for j in liste:
    indexnames.append(df2[df2['PassengerId']==j].index)

for k in indexnames:
    df2.drop(k, inplace=True)

df2
```

```python
#We set the passengerID as index then we create a new dataframe with all the values of the train dataset and the gender submission dataset

df_train = df3.set_index(['PassengerId'])
df_train.fillna(-1, inplace=True)

df1.set_index(['PassengerId'],inplace=True)
df2.set_index(['PassengerId'],inplace=True)
df_test = pd.merge(df1,df2,on=['PassengerId'])
df_test.fillna(-1, inplace=True)

df_train
```

```python
#this fuction print the percentage of values not null in the dataset
def nb_vall_null(name_column):
    nb_data=0
    for i in df_train[name_column]:
        filt = (i == -1)
        if filt == True:
            nb_data+=1
    return nb_data

nb_vall_null('Name')
```

```python
#this function print a boolean value who represent if yes or no all the values of a columns are unique
def unique(x):
    return (len(df_train[x].unique()) == 418-nb_vall_null(x))

for e in df_train.columns.to_list():
    print(f"{e} :{unique(e)}")
```

```python
#We delete all the column useless 
df_train.drop(['Name'],axis='columns' ,inplace=True) #all the names are unique and act like a label as the passengerID therefore this column is useless
df_test.drop(['Name'],axis='columns' ,inplace=True)
df_train.drop(['Cabin'],axis='columns' ,inplace=True) #the column cabin have not enough value therefore we decide to drop this column
df_test.drop(['Cabin'],axis='columns' ,inplace=True)
```

```python
df_encoded_test = pd.get_dummies(df_test, columns=['Sex','Ticket','Embarked'], drop_first=True)
df_encoded_test
```

```python
df_encoded_train = pd.get_dummies(df_train, columns=['Sex','Ticket','Embarked'], drop_first=True)
df_encoded_train
```

```python
def fill_column_lack():
    add= False
    list_to_add= []
    #fill columns of the test datatset with the columns of the train dataset not present in the test datatset
    for i in df_encoded_train.columns:
        for j in df_encoded_test.columns:
            if i==j:
                add=True
        if add == False:
            list_to_add.append(i)
        add=False
    for k in list_to_add:
        df_encoded_test[f'{k}'] = False
    add= False
    list_to_add= []
    #fill columns of the train datatset with the columns of the test dataset not present in the train dataset
    for i in df_encoded_test.columns:
        for j in df_encoded_train.columns:
            if i==j:
                add=True
        if add == False:
            list_to_add.append(i)
        add=False
    for k in list_to_add:
        df_encoded_train[f'{k}'] = False

fill_column_lack()
#we sort the index to have in the test and train dataset columns in the same order 
df_encoded_train.sort_index(axis = 1, ascending = True,inplace=True)
df_encoded_test.sort_index(axis = 1, ascending = True,inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,f1_score
from sklearn.preprocessing import StandardScaler
#apply of the logisic regression algorithm on the train dataset
def classification(X,Z):
    label = Z.values.reshape(-1,1)
    Train = LogisticRegression(C=0.5,solver='lbfgs',max_iter=102)
    Train.fit(X,label)
    return Train


train = classification(df_encoded_train.drop(['Survived'], axis='columns'),df_encoded_train['Survived'])
pred = train.predict(df_encoded_test.drop(['Survived'], axis='columns'))
```

```python
#display of the accuracy score
print(accuracy_score(np.array(df_encoded_test['Survived']),pred))
```

```python
col = np.array(df_encoded_test.drop(['Survived'], axis='columns').columns.tolist())
coef_df = pd.DataFrame(train.coef_, columns=col)
coef_df['intercept'] = train.intercept_
print(coef_df)
```
